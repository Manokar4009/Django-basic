```python
from bs4 import BeautifulSoup as bs
import requests 
import csv
import pandas as pd
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:98.0) Gecko/20100101 Firefox/98.0'
}
link = 'https://health.usnews.com/doctors/houston-tx'
page = requests.get(link, headers = headers)

print(page)
```

```python
#requesting to get a data from given link
page = requests.get(link, headers = headers)
```

```python
# checking status code
page.status_code
```

```python
# getting content from link
page.content
```

```python
soup = bs(page.content, 'html.parser')
#it gives us the visual representation of data
print(soup.prettify())
```

```python
name=soup.find('p', class_="Paragraph-sc-1iyax29-0 dUssLn")
print(name)
```

```python
a = soup.find_all('h2', class_="Heading-sc-1w5xk2o-0 hmwLHV mr3")
b = soup.find_all('span', class_="Span-sc-19wk4id-0 gQQdEn ml3")
c = soup.find_all('p', class_="Paragraph-sc-1iyax29-0 fJipda")
d = soup.find_all('p', class_="Paragraph-sc-1iyax29-0 lbGMQc")
e = soup.find_all('p', class_="Paragraph-sc-1iyax29-0 dUssLn")

Doctor_name = []
Reviews = []
category  = []
Hospital = []
Address = []

for title in a:
  Doctor_name.append(title.text.strip())
for title in b:
  Reviews.append(title.text.strip())
  if Reviews == '':
    a='NA'
    Reviews.append(a) 
  else :
    Reviews.append(title.text.strip())
for title in c:
 category.append(title.text.strip())
for title in d:
 Hospital.append(title.text.strip())
for title in e:
 Address.append(title.text.strip())

print(Doctor_name)
print(Reviews)
print(category)
print(Hospital)
print(Address)
```

```python
print(len(Doctor_name))
print(len(Reviews))
print(len(category))
print(len(Hospital))
print(len(Address))
```
